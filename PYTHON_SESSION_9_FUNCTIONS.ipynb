**Local variable and Global variable**

- Local variables means the variables inside the function call
- Global variables means the variables outside the function call
- The advantage of global variable is once you define the variables outside means, you can use those variables anywhere
- You can use global variable inside function also
- But you can not use local variables outside the function

```python
# Case-1
def mul():
    a=10
    b=20
    multiplication=a*b
    print(multiplication)
mul()
# a,b are local variables
# These values we can not use outside the function
```

```python
# Case-2
def mul():
    b=20
    multiplication=a*b
    print(multiplication)
a=10
mul()
# b is local variables
# a is global variables
# we can not use b outside the function
```

```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
def calculator():
    a=eval(input("Enter the value of a:"))
    b=eval(input("Enter the value of b:"))
    option=eval(input("Enter a option between 1-4"))
    if (option==1):
        add(a,b)
    elif (option==2):
        sub(a,b)
    elif (option==3):
        mul(a,b)
    elif (option==4):
        div(a,b)
    else:
        print("Enter a number from the given options")
calculator()
```
