```python
##################################################################################################################################
num1=10
num2=20
add=num1+num2
print(add)

#################################################################################################################################

num1=eval(input("Enter the number 1 value: "))
num2=eval(input("Enter the number 2 value: "))
add=num1+num2
print(add)

#################################################################################################################################

import random
num1=random.randint(1,20)
num2=random.randint(1,20)
add=num1+num2
print(f"Addition of {num1} and {num2} is {add}")

###################################################################################################################################

n1=10
n2=eval(input("Enter the n2 value:"))
n3=random.randint(1,20)
total=n1+n2+n3
avg=(n1+n2+n3)/3
print(f"Addition of {n1},{n2} and {n3} is {total}")
print(f"Average of {n1},{n2} and {n3} is {avg}")
```
